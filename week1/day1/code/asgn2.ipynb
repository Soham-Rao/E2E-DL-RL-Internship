```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'..\Datasets\misc\creditcard.csv')  
print(df['Class'].value_counts())
```

```python
non_fraud = df[df['Class'] == 0]['Amount']
fraud = df[df['Class'] == 1]['Amount']

plt.figure(figsize=(8, 5))
plt.hist([non_fraud, fraud], bins=50, label=['Non-Fraud', 'Fraud'], color=['blue', 'orange'], alpha=0.7, log=True)
plt.title('Transaction Amount Distribution (log-scaled)')
plt.xlabel('Transaction Amount')
plt.ylabel('Number of Transactions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df[df['Class'] == 0]['Time'], df[df['Class'] == 0]['Amount'], s=1, label='Non-Fraud', alpha=0.4)
plt.scatter(df[df['Class'] == 1]['Time'], df[df['Class'] == 1]['Amount'], s=10, label='Fraud', alpha=0.7, color='red')
plt.xlabel('Time (seconds since first transaction)')
plt.ylabel('Amount')
plt.title('Time vs Amount by Transaction Class')
plt.legend()
plt.grid(True)
plt.show()
```
